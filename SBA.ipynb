## Read the total profit of all months and show it using a Bar plot

```python
plt.bar(
    x=df['month_number'],
    height=df['total_profit']
)
plt.grid(linestyle='--', zorder=2)
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.title('Company profit per month')

plt.show()
```

## Line plot

```python
month_list = df['month_number'].tolist()
profit_list = df['total_profit'].tolist()

plt.figure(figsize=(8,5))

plt.plot(month_list,
         profit_list,
         color='red',
         marker='o',
         label='Profit data of last year',
         markerfacecolor='black',
         linestyle='--',
         linewidth='3')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.xticks(month_list)
plt.legend()
plt.show()
```

```python
facecream = df['facecream'].tolist()
facewash = df['facewash'].tolist()
toothpaste = df['toothpaste'].tolist()
bathingsoap = df['bathingsoap'].tolist()
shampoo = df['shampoo'].tolist()
moisturizer = df['moisturizer'].tolist()

plt.figure(figsize=(8,6))
plt.plot(month_list, facecream, color='blue', marker='o', markerfacecolor='blue', label='Face cream Sales Data')
plt.plot(month_list, facewash, color='orange', marker='o', markerfacecolor='orange', label='Face Wash Sales Data')
plt.plot(month_list, toothpaste, color='green', marker='o', markerfacecolor='green', label='Toothpaste Sales Data')
plt.plot(month_list, bathingsoap, color='red', marker='o', markerfacecolor='red', label='Bathingsoap Sales Data')
plt.plot(month_list, shampoo, color='purple', marker='o', markerfacecolor='purple', label='Shampoo Sales Data')
plt.plot(month_list, moisturizer, color='brown', marker='o', markerfacecolor='brown', label='Moisturizer Sales Data')
plt.xticks(month_list)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales Data')
plt.legend()
plt.show()
```

## Bathing Soap Scatter

```python
plt.scatter(month_list, bathingsoap, label='bathing soap Sales data')
plt.grid(linestyle='--', zorder=2)
plt.xticks(month_list)
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.title('bathsoap Sales data')
plt.legend()
plt.show()
```

```python
date=["25","26","27"]
temp=[8.5,10.5,6.8]
fahrenheit = [51, 47, 44]

plt.figure(figsize=(6,4))
plt.plot(date, temp, marker='o', markerfacecolor='blue', label='Celsius')
plt.plot(date, fahrenheit, marker='o', markerfacecolor='red', label='Fahrenheit')
plt.xlabel('Day of December')
plt.ylabel('Temperature')
plt.title('Date-wise Temperature')
plt.yticks()
plt.grid()
plt.legend()
plt.show()
```
